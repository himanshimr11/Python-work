```python
import pandas as pd
import matplotlib.pyplot as plt
url='/content/AAPL_2006-01-01_to_2018-01-01.csv'
df=pd.read_csv(url)
print(df.head())
# Calculate the median of the 'Close' prices
median_Close=df['Close'].median()
print(f"median value:-{median_Close}")
# Calculate the absolute deviations from the median
df['Deviation'] = abs(df['Close'] - median_Close)
print(f"abosulute deviation of median:-{df['Deviation']}")
# Calculate the mean deviation
mean_deviation = df['Deviation'].mean()
print(f"mean deviation:-{mean_deviation}")
# Create the line plot of 'Close' prices
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Close Price')

# Highlight deviations from the median
plt.fill_between(df.index, df['Close'], median_Close,
                 where=(df['Close'] > median_Close),
                 color='lightgreen', alpha=0.5, label='Above Median')
plt.fill_between(df.index, df['Close'], median_Close,
                 where=(df['Close'] < median_Close),
                 color='lightcoral', alpha=0.5, label='Below Median')

# Add horizontal line for the median
plt.axhline(median_Close, color='blue', linestyle='--', label='Median')

# Customize the plot
plt.title('Apple Close Prices with Deviations from Median')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
